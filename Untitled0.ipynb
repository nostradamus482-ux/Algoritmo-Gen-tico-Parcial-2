```python
# ==========================================================
#  Genetic Algorithm (GA) para detección de k-mers con LLR
#  Sin Biopython (lector FASTA nativo)
#  Ejecuta 3 corridas por semilla usando tus FASTA reales
# ==========================================================

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime

# -----------------------------
# Versiones (reproducibilidad)
# -----------------------------
print("NumPy:", np.__version__)
print("Pandas:", pd.__version__)
print("Matplotlib:", plt.matplotlib.__version__)

# -----------------------------
# Funciones auxiliares
# -----------------------------

def read_fasta_sequences(file_path):
    """Lee secuencias desde un archivo FASTA (sin usar Biopython)."""
    seqs = []
    seq = ""
    with open(file_path, "r") as f:
        for line in f:
            if line.startswith(">"):
                if seq:
                    seqs.append(seq)
                    seq = ""
            else:
                seq += line.strip().upper()
        if seq:
            seqs.append(seq)
    return seqs

def kmer_counts(seqs, k):
    """Cuenta todos los k-mers de una lista de secuencias."""
    counts = {}
    for seq in seqs:
        for i in range(len(seq) - k + 1):
            kmer = seq[i:i+k]
            if set(kmer) <= {"A","C","G","T"}:
                counts[kmer] = counts.get(kmer, 0) + 1
    return counts

def compute_llr(pos_counts, neg_counts, k):
    """Calcula el LLR (log-likelihood ratio) entre dos conjuntos de k-mers."""
    all_kmers = set(pos_counts) | set(neg_counts)
    llr = {}
    for kmer in all_kmers:
        n_pos = pos_counts.get(kmer, 0) + 1  # pseudocuenta
        n_neg = neg_counts.get(kmer, 0) + 1
        p_pos = n_pos / sum(pos_counts.values())
        p_neg = n_neg / sum(neg_counts.values())
        llr[kmer] = np.log2(p_pos / p_neg)
    return llr

def fitness_function(kmer, llr_dict):
    """Devuelve el valor de fitness del k-mer basado en el LLR."""
    return llr_dict.get(kmer, 0)

def initialize_population(k, pop_size, rng):
    """Genera una población inicial aleatoria de k-mers."""
    alphabet = ["A", "C", "G", "T"]
    return ["".join(rng.choice(alphabet, size=k)) for _ in range(pop_size)]

def mutate(kmer, p_mut, rng):
    """Muta un k-mer con probabilidad p_mut por posición."""
    alphabet = ["A", "C", "G", "T"]
    kmer_list = list(kmer)
    for i in range(len(kmer_list)):
        if rng.random() < p_mut:
            kmer_list[i] = rng.choice([b for b in alphabet if b != kmer_list[i]])
    return "".join(kmer_list)

def crossover(parent1, parent2, rng):
    """Crossover de un punto entre dos padres."""
    point = rng.integers(1, len(parent1))
    return parent1[:point] + parent2[point:]

def tournament_selection(pop, fitnesses, rng, t_size=3):
    """Selección por torneo."""
    selected = rng.choice(len(pop), size=t_size, replace=False)
    best_idx = selected[np.argmax([fitnesses[i] for i in selected])]
    return pop[best_idx]

# -----------------------------
# Algoritmo Genético Principal
# -----------------------------
def run_ga_llr(pos_seqs, neg_seqs, k, params, seed):
    rng = np.random.default_rng(seed)

    pos_counts = kmer_counts(pos_seqs, k)
    neg_counts = kmer_counts(neg_seqs, k)
    llr_dict = compute_llr(pos_counts, neg_counts, k)

    pop = initialize_population(k, params["pop_size"], rng)
    best_history = []
    stagnant_gens = 0
    last_best = -1e9

    for gen in range(params["max_gens"]):
        fitnesses = [fitness_function(k, llr_dict) for k in pop]
        best_idx = np.argmax(fitnesses)
        best_fit = fitnesses[best_idx]
        mean_fit = np.mean(fitnesses)
        std_fit = np.std(fitnesses)
        best_history.append((gen, best_fit, mean_fit, std_fit))

        # criterio de paro
        if abs(best_fit - last_best) < 1e-3:
            stagnant_gens += 1
        else:
            stagnant_gens = 0
        last_best = best_fit
        if stagnant_gens >= 30:
            break

        # Nueva población
        new_pop = []
        elites = np.argsort(fitnesses)[-params["elitism"]:]
        for e in elites:
            new_pop.append(pop[e])

        while len(new_pop) < params["pop_size"]:
            p1 = tournament_selection(pop, fitnesses, rng)
            p2 = tournament_selection(pop, fitnesses, rng)
            if rng.random() < params["p_crossover"]:
                child = crossover(p1, p2, rng)
            else:
                child = p1
            child = mutate(child, params["p_mut"], rng)
            new_pop.append(child)
        pop = new_pop

    return best_history, pop[np.argmax(fitnesses)], max(fitnesses)

# -----------------------------
# Parámetros del GA
# -----------------------------
params = {
    "k": 8,
    "pop_size": 100,
    "max_gens": 200,
    "p_crossover": 0.8,
    "p_mut": 0.1,
    "elitism": 2
}

seeds = [42, 101, 2025]
runs_per_seed = 3

# -----------------------------
# Cargar FASTA reales
# -----------------------------
pos_seqs = read_fasta_sequences("promoters_pos.fasta")
neg_seqs = read_fasta_sequences("promoters_neg.fasta")

# -----------------------------
# Ejecutar GA (3 corridas por semilla)
# -----------------------------
records = []
summary = []

for seed in seeds:
    rng_seed = np.random.default_rng(seed)
    for run in range(runs_per_seed):
        subseed = rng_seed.integers(1e9)
        best_history, best_kmer, best_fit = run_ga_llr(pos_seqs, neg_seqs, params["k"], params, subseed)
        for gen, b, m, s in best_history:
            records.append({
                "seed": seed,
                "subrun": run+1,
                "generation": gen,
                "best_fitness": b,
                "mean_fitness": m,
                "std_fitness": s
            })
        summary.append({
            "seed": seed,
            "subrun": run+1,
            "best_kmer": best_kmer,
            "best_fitness": best_fit,
            "gens_executed": len(best_history)
        })

# -----------------------------
# Guardar resultados
# -----------------------------
df = pd.DataFrame(records)
df.to_csv("GA_results_extended.csv", index=False)

meta = pd.DataFrame(summary)
meta.to_csv("GA_summary_extended.csv", index=False)

# -----------------------------
# Graficar convergencia
# -----------------------------
plt.figure(figsize=(8,5))
for seed in seeds:
    subset = df[df["seed"] == seed]
    grouped = subset.groupby("generation").agg({"best_fitness": ["mean","std"]})
    x = grouped.index
    y = grouped["best_fitness"]["mean"]
    e = grouped["best_fitness"]["std"]
    plt.plot(x, y, label=f"Seed {seed}")
    plt.fill_between(x, y-e, y+e, alpha=0.2)
plt.xlabel("Generación")
plt.ylabel("Fitness (LLR)")
plt.title("Convergencia del GA (3 corridas por semilla)")
plt.legend()
plt.tight_layout()
plt.savefig("GA_convergence_extended.png", dpi=300)
plt.show()

# -----------------------------
# Guardar metadatos
# -----------------------------
with open("GA_metadata_extended.txt", "w") as f:
    f.write("GA LLR — 3 corridas por semilla\n")
    f.write(f"Fecha: {datetime.now()}\n\n")
    f.write(str(params) + "\n\n")
    f.write(pd.DataFrame(summary).to_string(index=False))

print("✅ Archivos generados:")
print(" - GA_results_extended.csv")
print(" - GA_summary_extended.csv")
print(" - GA_convergence_extended.png")
print(" - GA_metadata_extended.txt")
```
